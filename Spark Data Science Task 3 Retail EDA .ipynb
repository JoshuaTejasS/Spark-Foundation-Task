```python
store = data.drop(['Postal Code'], axis=1)       #Dropping 'Postal code' as it is numerical and not necessary for analysis
```

```python
correlation = store.corr()                       #Correlation
```

```python
correlation
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)  #Correlation Heatmap
```

```python
covariance = store.cov()                         #Covariance
```

```python
covariance
```

```python
store['State'].value_counts()                   #Counts No. of data from each state
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=store['State'])              #Visual Representation of Purchases in Each State
plt.xticks(rotation=90)
plt.title("States")
plt.show()
```

```python
#Inference: The Purchases are highest in California and least in Wyoming
```

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(20,15))
sns.barplot(x='Sub-Category',y='Profit', data=store, palette='PuBuGn')
plt.suptitle('Profits on Each Category', fontsize=16)
plt.show()
```

```python
#Inference: Profit is high for Copiers and least for Fasteners while the Loss is maximum for Tables and minimum for Supplies.
```

```python
figsize=(15,10)
sns.pairplot(store,hue='Sub-Category')   #Pairplot with respect to Sub-Categories
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(store,hue='Category')      #Pairplot with respect to Categories
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(store,hue='Segment')        #Pairplot with respect to Segments
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(store,hue='Region')               #Pairplot with respect to Regions
plt.show()
```

```python
store['Region'].value_counts()
```

```python
plt.pie(store['Region'].value_counts(),labels=["West","East","Central","South"])   #Representing Purchases from Regions
plt.show()
```

```python
#Inference: No. of Purchases is Maximum at the West Region and Minimum at the South Region
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='Discount',y='Sales', data=store , color='b',label='Discount vs Sales')    #Line chart for Discount vs Sales
plt.legend()
plt.show()
```

```python
#Inference: Sales is at its peak when items sold for a discount of 0.5 (50%)
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='Discount',y='Profit', data=store , color='g',label='Discount vs Profit')   #Line chart for discount vs profit
plt.legend()
plt.show()
```

```python
#Inference: The Profit is at its peak when items sold for a discount of 0.1 (10%) and there is no profit when items are sold at 25% and higher discounts
```

```python
#Final Inference: It is most effective to sell products at 10-20% discount as it would yield a profit and also keep the sales going.
```
